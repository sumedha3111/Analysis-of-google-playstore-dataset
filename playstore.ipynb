```python
#dropping the rows with null values
clean_df = df.dropna(how='any')
```

```python
#shape of the dataframe before and after dropping rows
print ("new dataframe shape: ", clean_df.shape)
print ("old dataframe shape: ", df.shape)
```

```python
#number of duplicate rows
duplicate_rows_clean_df = clean_df[clean_df.duplicated()]
print ("number of duplicate rows: ", duplicate_rows_clean_df.shape)
```

```python
#dropping the rows with duplicate values
print ("shape of dataframe after dropping duplicates: ", clean_df.drop_duplicates().shape) 
```

```python
#count plot for number of free apps and paid apps
sns.set(style="darkgrid")
a1 = sns.countplot(x="Type", data=clean_df,palette="Set3")
```

```python
#barplot for number of installs and rating
ax = sns.barplot(x="Rating", y="Installs", data=clean_df,
                 palette="Blues_d")
```

```python
#barplot for category and rating
barfig, ax = plt.subplots(figsize=(10,10))
sns.barplot(x = "Rating", y = "Category", ax=ax, data=clean_df)
```

```python
#countplot for categories
sns.set(style="darkgrid")
fig, ax = plt.subplots(figsize=(7,10))
a1 = sns.countplot(y="Category", ax=ax, data=clean_df,palette="Set3")
```

```python
#countplots for content rating
sns.set(style="darkgrid")
fig, ax = plt.subplots(figsize=(10,5))
a1 = sns.countplot(x="Content Rating", ax=ax, data=clean_df,palette="Set3")
```

```python
#countplot for genres
sns.set(style="darkgrid")
fig, ax = plt.subplots(figsize=(7,25))
a1 = sns.countplot(y="Genres", ax=ax, data=clean_df,palette="Set3")
```

```python
#countplot for android versions
sns.set(style="darkgrid")
fig, ax = plt.subplots(figsize=(5,6))
a1 = sns.countplot(y="Android Ver", ax=ax, data=clean_df,palette="Set3")
```

```python
sns.distplot(clean_df.Rating)
```

```python
#correlation matrix
clean_df.dtypes
clean_df["Type"] = (clean_df["Type"] == "Paid").astype(int)
fig = plt.figure(figsize = (9, 9))
corr = clean_df.apply(lambda x: x.factorize()[0]).corr()
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns,annot=True)
```
